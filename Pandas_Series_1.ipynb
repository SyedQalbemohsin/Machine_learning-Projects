```python
import numpy as np
import pandas as pd

```

```python
#@title Series from List
country = ['India','Pakistan','Nepal','USA']
print(pd.Series(country))
```

```python
runs = [10,20,25,45,65,45]
print(pd.Series(runs))
```

```python
#Custom index
marks = [55,65,75,85]
subjects = ['maths','english','urdu','oops']
print(pd.Series(marks,index=subjects))
# here we will get subjects instead of index
```

```python
#setting a name to the object
print(pd.Series(marks,index=subjects,name='Subjects_marks'))
```

```python
#@title Series from Dictionary
marks = {
    'maths':75,
    'english':55,
    'urdu':65,
    'oops':85
}
marks_series= pd.Series(marks,name='Subject_marks')
print(marks_series)
```
