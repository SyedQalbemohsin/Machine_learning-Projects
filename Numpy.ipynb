```python
# @title Broadcasting
#Broadcasting is a term describing how Numpy treats to different shapes of arrays during arithematic operation
# the smaller array is broadcasted to larger array so that they have compatible shapes

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)
print(a)
print(b)

print(a + b)
```

```python
# @title Broadcasting Rules
# 1. add 1 to the head of smaller array to make it same dimension like the larger array a[3,3,3] so if b[3] so make b[1,1,3]
# 2. then make 1 similar value the larger array have like b to [3,3,3]
# 3. error if there is no 1 dimension
```

```python
#example
a = np.arange(12).reshape(4,3)
b = np.arange(3)
print(a)
print(b)
print(a+b)
#this work because added 1 to head of b then make it (4,3) then added
```

```python
#example
a = np.arange(12).reshape(3,4)
b = np.arange(3)
print(a)
print(b)
print(a+b)
# we can only increase 1 to the value of larger array value
#this will not work because added 1 to head of b and the second value is 3 so we cant make it 4 in 1Dimension we make only first value to first value of larger array
```

```python
#example
a = np.arange(3).reshape(3,1)
b = np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
#this work because we only have to match both array values then added here because wwe are only increasing 1 not any other number
```

```python
#example
a = np.arange(4).reshape(2,2)
b = np.arange(1)
print(a)
print(b)
print(a+b)
#first 1 added to head of b then we increases 1 to the values of larger array
```

```python
#example
a = np.arange(12).reshape(4,3)
b = np.arange(12).reshape(3,4)
print(a)
print(b)
print(a+b)
# error because we dont have 1 on either side to increase
```
